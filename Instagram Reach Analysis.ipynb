## Analysis of Instagram Reach

```python
plt.figure(figsize=(10, 8))
plt.style.use('fivethirtyeight')
plt.title("Distribution of Impressions From Home")
sns.distplot(data['From Home'])
plt.show()
```

#### The impressions derived from the home section on Instagram provide a metric for seeing the extent to which my posts resonate with my followers. Analyzing these impressions, it becomes evident that achieving daily outreach to all followers is a challenge.

```python
plt.figure(figsize=(10, 8))
plt.title("Distribution of Impressions From Hashtags")
sns.distplot(data['From Hashtags'])
plt.show()
```

#### Hashtags are tools we use to categorize our posts on Instagram so that we can reach more people based on the kind of content we are creating. Looking at hashtag impressions shows that not all posts can be reached using hashtags, but many new users can be reached from hashtags.

```python
plt.figure(figsize=(10, 8))
plt.title("Distribution of Impressions From Explore")
sns.distplot(data['From Explore'])
plt.show()
```

#### The explore section is a recommendation system of Instagram. It recommends posts to platform users based on their preferences and interests. Based on the impressions I have received from the explore section, I can say that Instagram does not recommend our posts as much to the users. Some posts have received a good reach from the explore section, but itâ€™s still very low compared to the reach I receive from hashtags.

```python
home = data["From Home"].sum()
hashtags = data["From Hashtags"].sum()
explore = data["From Explore"].sum()
other = data["From Other"].sum()

labels = ['From Home','From Hashtags','From Explore','Other']
values = [home, hashtags, explore, other]

fig = px.pie(data, values=values, names=labels, 
             title='Impressions on Instagram Posts From Various Sources', hole=0.5)
fig.show()
```

#### The doughnut chart shows that almost 50% of the reach is from my followers, 33.6% is from hashtags, 19.2% is from the explore section, and 3.05% is from other sources.

## Analysis of Relationships

```python
correlation = data.corr()
print(correlation["Impressions"].sort_values(ascending=False))
```

## Analysis of Conversion Rate
#### On Instagram, the conversation rate is defined by how many followers you are getting from the number of profile visits from a post. The formula that you can use to calculate conversion rate is (Follows/Profile Visits) * 100. 

```python
conversion_rate = (data["Follows"].sum() / data["Profile Visits"].sum()) * 100
print(conversion_rate)
```

#### The conversation rate of my Instagram account is 31% which sounds like a reasonable conversation rate.

```python
figure = px.scatter(data_frame = data, x="Profile Visits",
                    y="Follows", size="Follows", trendline="ols", 
                    title = "Relationship Between Profile Visits and Followers Gained")
figure.show()
```

#### The relationship between profile visits and followers gained is also linear.

## Instagram Reach Prediction Model

#### I will train a machine learning model to predict the reach of an Instagram post based on features such as likes, saves, comments, shares, profile visits, and follows.
#### I will carry out a train test split where you split your data into a training set and a testing set. The training set is used for training the model, and the unseen testing set is used to test your accuracy of the model.

```python
x = np.array(data[['Likes', 'Saves', 'Comments', 'Shares', 
                   'Profile Visits', 'Follows']])
y = np.array(data["Impressions"])
xtrain, xtest, ytrain, ytest = train_test_split(x, y, 
                                                test_size=0.2, 
                                                random_state=42)
```

```python
model = PassiveAggressiveRegressor()
model.fit(xtrain, ytrain)
model.score(xtest, ytest)
```

#### Now letâ€™s predict the reach of an Instagram post by giving inputs to the machine learning model:

```python
# Features = [['Likes','Saves', 'Comments', 'Shares', 'Profile Visits', 'Follows']]
features = np.array([[282.0, 233.0, 4.0, 9.0, 165.0, 54.0]])
model.predict(features)
```

#### This means that, according to the trained model, given the specified amounts of likes, saves, comments, shares, profile visits, and follows for an Instagram post, the predicted reach is approximately 10320.6788364.
